Remove Constant Columns

```python
wb.columns
```

```python
len(wb['f1'].unique())==1 #checking if the column has only one unique value (constant column)
```

```python
def drop_constant_columns(dataframe): #defining a function that drops constant value columns of pandas dataframe
    result = dataframe.copy() #the result will be the copy of the original dataframe with no constant columns
    
    for column in dataframe.columns:
        if(len(dataframe[column].unique()) == 1 ): #if length of the vector created by unique values in column is 1 (constant column)
            print(column) #print me the name of that column
            result = result.drop(column,axis=1) # remove that column from the data
    return result        
```

```python
wb=drop_constant_columns(wb)
```

```python
Q_constant = get_constant_features(wb, threshold=0.7) 
# defining a varaiable that will store all quasi constant features from wb
# we chose threshold of 70%, but you can choose anything, depending on the data
```

```python
Q_constant.shape
```

```python
name = Q_constant[Q_constant.Desc=="Quasi Constant"].Var #storing all data with "Desc" column = "Quasi COnstant" in variable "name"
```

```python
wb  = wb.drop(name, axis=1) #removing Quasi constants
```

Drop Correlated

```python
correlation_matrix = wb.corr()
```

```python
correlation_matrix
```

```python
correlated_features = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.8:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)
```

```python
len(correlated_features)
```

```python
wb.drop(correlated_features,axis=1,inplace=True)
```
